```python
with open('o.txt', 'r') as f:
    data = f.readlines()
data = [i.split() for i in data]
for i in range(len(data)):
    for j in range(len(data[i])):
        data[i][j] = int(data[i][j])
        
diskReadsLRU = data[9]
diskReadsClock = data[10]
diskReadsMRU = data[11]
```

```python
import matplotlib.pyplot as plt

plt.plot(diskReadsLRU, label='LRU')
plt.plot(diskReadsClock, label='Clock')
plt.plot(diskReadsMRU, label='MRU')
plt.legend()
plt.title('Disk Reads of Replacement algorithms for Join Query')

plt.show()
```
