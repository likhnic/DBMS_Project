```python
with open('o.txt', 'r') as f:
    data = f.readlines()
data = [i.split() for i in data]
for i in range(len(data)):
    for j in range(len(data[i])):
        data[i][j] = int(data[i][j])
        
diskReadsLRU = data[12]
diskReadsMRU = data[13]
diskReadsClock = data[14]
```

```python
import matplotlib.pyplot as plt
frames = [i for i in range(2, 41)]
plt.plot(frames, diskReadsLRU, label='LRU')
plt.plot(frames, diskReadsMRU, label='MRU')
plt.plot(frames, diskReadsClock, label='Clock')
plt.scatter(frames, diskReadsLRU)
plt.scatter(frames, diskReadsMRU)
plt.scatter(frames, diskReadsClock)
plt.xlabel('Number of Frames in Buffer Pool')
plt.ylabel('Number of Disk Reads')
plt.legend()
plt.title('Disk Reads for Join Query')

plt.show()
```
